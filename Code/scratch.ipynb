```python
from Bio import PDB
```

```python
pdb_file = '../Data/pdbs/1ci0.pdb'
pdb_io = PDB.PDBIO()
pdb_parser = PDB.PDBParser()
clean_pdb_file = pdb_file.replace('.pdb', '.clean.pdb')
structure = pdb_parser.get_structure(" ", pdb_file)
chain_dict = {}
models = 0
for model in structure:
    models += 1
    for chain in model:
        new_number = 1
        for i, residue in enumerate(chain.get_residues()):
            res_id = list(residue.id)
            if res_id[1] != new_number:
                res_id[1] = new_number
                residue.id = tuple(res_id)
            new_number+=1
        chain_dict[chain.id] = new_number
if models != 1:
    print('More than one model in structure file, behavior of script unknown')
    
    
print(chain_dict)
chain_to_keep = sorted(chain_dict.items(), key=lambda x: x[1])
chain_to_keep = chain_to_keep[-1][0]
print(chain_to_keep)

for model in structure:
    for chain in model:
        if chain.id != chain_to_keep:
            model.detach_child(chain.id)

print('###')
for model in structure:
    print(model.id)
    for chain in model:
        print(chain.id)
pdb_io.set_structure(structure)
pdb_io.save(clean_pdb_file)
```

```python
all_records = []
pdb_files = list(glob.glob('../../Structural_analysis/ecoli_all_pdbs/PDBs/*.clean.pdb'))
for i, fasta_file in enumerate(glob.glob('../../Structural_analysis/ecoli_all_pdbs/wt_fastas/*.fasta')):
    if fasta_file.replace('/wt_fastas/', '/PDBs/').replace('.fasta', '.clean.pdb') in pdb_files:
        print(fasta_file)
        print(fasta_file.split('/')[-1].replace('.fasta', ''))
        records = list(SeqIO.parse(fasta_file, 'fasta'))
        assert len(records) == 1
        records[0].id = fasta_file.split('/')[-1].replace('.fasta', '')
        all_records.extend(records)
print(len(all_records))
```

```python
import numpy as np

def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
```
