```python
def create_markov_chain():
    markov_chain = {}
    for from_base in 'ATGC':
        # Generate random transition probabilities by dividing
        # [0,1] into four intervals of random length
        slice_points = sorted(
           [0] + [random.random()for i in range(3)] + [1])
        transition_probabilities = \
           [slice_points[i+1] - slice_points[i] for i in range(4)]
        markov_chain[from_base] = {base: p for base, p
                         in zip('ATGC', transition_probabilities)}
    return markov_chain

mc = create_markov_chain()
print(mc)
print(mc['A']['T'])
```

```python
kappa = 1
dicty = {'A' : {'T':(1/3)/((2/3)+(1/3*kappa)), 'G':(1/3)*kappa/((2/3)+(1/3*kappa)), 'C':(1/3)/((2/3)+(1/3*kappa))},
         'T' : {'A':(1/3)/((2/3)+(1/3*kappa)), 'G':(1/3)/((2/3)+(1/3*kappa)), 'C':(1/3)*kappa/((2/3)+(1/3*kappa))},
         'G' : {'A':(1/3)*kappa/((2/3)+(1/3*kappa)), 'T':(1/3)/((2/3)+(1/3*kappa)), 'C':(1/3)/((2/3)+(1/3*kappa))},
         'C' : {'A':(1/3)/((2/3)+(1/3*kappa)), 'T':(1/3)*kappa/((2/3)+(1/3*kappa)), 'G':(1/3)/((2/3)+(1/3*kappa))}}

new_dicty = {}
for from_key in dicty:
    new_dicty[from_key] = {}
    last = 0
    for to_key, val in dicty[from_key].items():
        new_dicty[from_key][to_key] = (last, last + dicty[from_key][to_key])
        last = last + dicty[from_key][to_key]
```

```python
new_dicty
```

```python
starting_seq_list = [random.choice(['A', 'T', 'G', 'C']) for i in range(100)]
for i in range(1000):
    nt_index = random.randint(0, len(starting_seq_list)-1)
    nt = starting_seq_list[nt_index]
    rand_number = random.random()
    for new_nt in new_dicty[nt].keys():
        if rand_number >= new_dicty[nt][new_nt][0]:
            if rand_number < new_dicty[nt][new_nt][1]:
                starting_seq_list[nt_index] = new_nt
                break
    print(Counter(starting_seq_list))
```
